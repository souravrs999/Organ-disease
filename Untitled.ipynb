```python
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
from matplotlib.cm import rainbow
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#cardiac -df_heart, lungs -df_lungs, liver -df_liver, kidney -df_kidney
```

```python
#heart_dataframe
combined_csv = pd.concat([pd.read_csv(f) for f in glob.glob('./heart/*.csv')])
df_heart = pd.DataFrame(combined_csv)
#lungs_dataframe
df_lungs = pd.DataFrame(pd.read_csv('./lungs/lung_cancer_examples.csv'))
df_lungs = df_lungs.drop(['Name','Surname'],axis=1)
# kidney_dataframe
df_kidney = pd.DataFrame(pd.read_csv('./kidney/kidney_disease.csv'))
# liver_dataframe
df_liver = pd.DataFrame(pd.read_csv('./liver/indian_liver_patient.csv'))
```

# HEART

```python
df_heart.head()
```

```python
corrmat = df_heart.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
map = sns.heatmap(df_heart[top_corr_features].corr(),annot=True,cmap='RdYlGn')
```

```python
dataset = pd.get_dummies(df_heart, columns=['sex','cp','fbs','restecg','exang','slope','ca','thal'])
```

```python
cts = ['age','trestbps','chol','thalach','oldpeak']
```

```python
dataset[cts] = StandardScaler().fit_transform(dataset[cts])
```

```python
dataset.head()
```

```python
X = dataset.drop(['target'],axis=1)
y = dataset['target']
```

```python
knn_scores = []
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score=cross_val_score(knn_classifier,X,y,cv=10)
    knn_scores.append(score.mean())
```

```python
plt.figure(figsize=(20,20))
plt.plot([k for k in range(1, 21)], knn_scores, color = 'green')
for i in range(1,21):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=1)
score = cross_val_score(knn_classifier,X,y,cv=10).mean()
score
```

```python
rf_scores = []
n_estimators = [1,2,4,8,16,32,64,100,200]
for k in n_estimators:
    randomforest_classifier= RandomForestClassifier(n_estimators=k)
    score=cross_val_score(randomforest_classifier,X,y,cv=10).mean()
    rf_scores.append(score)
```

```python
plt.figure(figsize=(20,20))
plt.plot([k for k in n_estimators], rf_scores, color = 'green')
for i in n_estimators:
    plt.text(i, rf_scores[n_estimators.index(i)], (i, rf_scores[n_estimators.index(i)]))
plt.xticks([i for i in [1,2,4,8,16,32,64,100,200]])
plt.xlabel('Number of estimators (K)')
plt.ylabel('Scores')
plt.title('RandomForest Classifier scores for different K values')
```

```python
randomforest_classifier= RandomForestClassifier(n_estimators=32)
score=cross_val_score(randomforest_classifier,X,y,cv=10).mean()
score
```
